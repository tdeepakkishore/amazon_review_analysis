```python
# Libraries
import json
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing Data
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# Data Head
movies.head(3)
```

```python
# Credits Head
credits.head(3)
```

```python
# Movies Shape
movies.shape
```

```python
# Credits Shape
credits.shape
```

```python
# Merging Movies and Credits
movies = movies.merge(credits, on ='title')
```

```python
# Data Columns
movies.columns
```

```python
# Selecting Required Columns
movies = movies.loc[:,['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
# Data Head
movies.head()
```

```python
# Checking for Null Values
movies.isnull().sum()
```

```python
# Dropping Null Values
movies.dropna(inplace=True)
```

```python
# Checking for Null Values
movies.isnull().sum()
```

```python
# Searching for Duplicate Values
movies.duplicated().sum()
```

```python
# Function to Extract Feature
def convert(obj):
    lst = [i['name'] for i in json.loads(obj)]
    return lst
```

```python
# Extracting Feature
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# Function to Extract Feature
def convert_cast(obj):
    cnt = 0
    lst = []
    for i in json.loads(obj):
        if cnt != 3:
            lst.append(i['name'])
            cnt += 1
    return lst
```

```python
# Extracting Feature
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
# Function to Fetch Director from Crew
def fetch_director(obj):
    lst = []
    for i in json.loads(obj):
        if i['job'] == 'Director':
            lst.append(i['name'])
    return lst
```

```python
# Fetching Director from Crew
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# Splitting the words in Overview
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# Removing Spaces between Text in Data
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
# Creating Feature using Other Features
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# Joining all Tags
movies['tags'] = movies['tags'].apply(lambda x:' '.join(x))
```

```python
# Converting Tags to Lowercase
movies['tags'] = movies['tags'].apply(lambda x:x.lower())
```

```python
# Data Head
movies.head()
```

```python
# Feature Selection
df = movies[['movie_id','title','tags']]
```

```python
# Data Head
df.head()
```

```python
# Stemming of Data
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Defining Function for Stemming
def stemmer_data(data):
    lst = []
    for i in data.split():
        lst.append(ps.stem(i))
    return ' '.join(lst)
```

```python
# Applying Function for Stemming
df['tags'] = df['tags'].apply(stemmer_data)
```

```python
# Data Head
df.head()
```

```python
# Vectorization of Data
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
# Converting Vectors to Array
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
# Vector Data
vectors[0]
```

```python
# Features Names in Tags
cv.get_feature_names()
```

```python
# Cosine Similarity to Find Relation Between Different Movies
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
# Defining Function for Finding Top 5 Similar Movies with Selected Movies
def recommend(movie):
    ind = df[df['title'] == movie].index[0]
    corr = similarity[ind]
    mov = sorted(list(enumerate(corr)),reverse=True,key=lambda x:x[1])[1:6]
    
    vals = [i[0] for i in mov]
    
    for i in df.iloc[vals,1].values:
        print(i)
```

```python
# Finding Similar Movies
recommend('John Carter')
```

```python
# Pickling File for Api Building
import pickle

pickle.dump(similarity,open('similarity.pkl','wb'))
pickle.dump(df.to_dict(),open('movies_dict.pkl','wb'))
```
